```python
from keras.datasets import mnist
from keras.models import Sequential
from keras.layers import Conv2D
from keras.layers import MaxPool2D
from keras.layers import Flatten
from keras.layers import Dense, Dropout, BatchNormalization
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python

train_dir = '/content/Gender/Taining'
validation_dir = '/content/Gender/Validation'


image_size = (100, 100)
batch_size = 32


train_datagen = ImageDataGenerator(
    rescale=1./255,
    #horizontal_flip=True
)


train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=image_size,
    batch_size=batch_size,
    class_mode='binary'
)


validation_datagen = ImageDataGenerator(rescale=1./255)


validation_generator = validation_datagen.flow_from_directory(
    validation_dir,
    target_size=image_size,
    batch_size=batch_size,
    class_mode='binary'
)
```

```python
#defining model
model = Sequential()
#adding convolution layer and pooling layer
model.add(Conv2D(32, (3, 3), activation='relu', input_shape=(100, 100, 3)))
model.add(MaxPool2D((2, 2)))
model.add(BatchNormalization())
model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(MaxPool2D((2, 2)))
model.add(Conv2D(128, (3, 3), activation='relu'))
model.add(MaxPool2D((2, 2)))
model.add(Conv2D(128, (3, 3), activation='relu'))
model.add(MaxPool2D((2, 2)))
model.add(Conv2D(128, (3, 3), activation='relu'))
model.add(MaxPool2D((2, 2)))
#adding fully connected layer
model.add(Flatten())
model.add(Dense(400, activation='relu'))
model.add(Dropout(0.25))
model.add(Dense(150, activation='relu'))
model.add(Dense(100, activation='relu'))
#adding output layer
model.add(Dense(2, activation='sigmoid'))
# Compile
model.compile(loss='sparse_categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model.fit(train_generator, epochs=10, batch_size=32, validation_data=validation_generator)
```

```python
loss, accuracy = model.evaluate(validation_generator)
#print(f'Accuracy: {accuracy*100}%')
```
